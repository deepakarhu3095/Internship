```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://en.wikipedia.org/wiki/Main_Page')
```

```python
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
header = soup.find_all(['h1', 'h2','h3','h4','h5','h6'])
header
```

```python
h = []
for i in header:
    h.append(i.text.replace("\n",""))
h    
```
